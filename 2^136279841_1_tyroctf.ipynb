```python
import sympy

def find_1024_bit_prime():
    # Start with the smallest 1024-bit number
    start = 2**1023 + 11436
    # Generate 1024-bit primes
    for p in sympy.primerange(start, 2**1024):
        # Calculate the expression
        expression = p**2 + 3*p + 1
        # Check if the expression is prime
        if sympy.isprime(expression):
            return p, expression
    return None, None

# Find the 1024-bit prime p and check the expression
p, result = find_1024_bit_prime()

if p is not None:
    print(f"1024-bit prime p: {p}")
    print(f"p^2 + 3p + 1: {result} (also prime)")
else:
    print("No suitable 1024-bit prime found.")
```
